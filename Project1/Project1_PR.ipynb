```python
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.data import loadlocal_mnist


X, y = loadlocal_mnist(
    images_path='train-images-idx3-ubyte',
    labels_path='train-labels-idx1-ubyte')
X_test, y_test = loadlocal_mnist(
    images_path='t10k-images-idx3-ubyte',
    labels_path='t10k-labels-idx1-ubyte')

print('Dimensions: %s x %s' % (X.shape[0], X.shape[1]))

train_features_images = X.reshape(X.shape[0],28,28)

np.amax(train_features_images[0])

def show_images(features_images,labels,start, howmany):
    for i in range(start, start+howmany):
        plt.figure(i)
        plt.imshow(features_images[i], cmap=plt.get_cmap('gray'))
        plt.title(labels[i])
    plt.show()
show_images(train_features_images, y, 1, 20)

plt.imshow(train_features_images[2], cmap=plt.get_cmap('gray'))
plt.show()

item_storage = dict()
for i in range(0,10):
    item_storage[i] = np.where(y==i)
print(item_storage)

mean_array = np.zeros((10,28,28))
for i in range(0,10):
    sum_temp = np.zeros((28,28))
    for z in item_storage[i]:
        for j in z:
            for k in range(0,28):
                 for l in range(0,28):
                        sum_temp[k][l]+= train_features_images[j][k][l]
    for k in range(0,28):
        for l in range(0,28):
            sum_temp[k][l]=sum_temp[k][l]/len(item_storage[i][0])
    mean_array[i] = sum_temp

plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/mean_0.png", mean_array[0])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/mean_1.png", mean_array[1])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/mean_2.png", mean_array[2])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/mean_3.png", mean_array[3])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/mean_4.png", mean_array[4])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/mean_5.png", mean_array[5])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/mean_6.png", mean_array[6])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/mean_7.png", mean_array[7])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/mean_8.png", mean_array[8])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/mean_9.png", mean_array[9])

std_array = np.zeros((10,28,28))
variance_arr = np.zeros((10,784))
for i in range(0,10):
    sum_temp = np.zeros((28,28))
    for z in item_storage[i]:
        for j in z:
            for k in range(0,28):
                 for l in range(0,28):
                        sum_temp[k][l] += np.power((train_features_images[j][k][l] - mean_array[i][k][l]),2)
    for k in range(0,28):
        for l in range(0,28):
            sum_temp[k][l]=np.sqrt(sum_temp[k][l]/len(item_storage[i][0]))
    std_array[i] = sum_temp
    variance_arr[i] = (sum_temp*sum_temp).ravel()

plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/std_0.png", std_array[0])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/std_1.png", std_array[1])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/std_2.png", std_array[2])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/std_3.png", std_array[3])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/std_4.png", std_array[4])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/std_5.png", std_array[5])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/std_6.png", std_array[6])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/std_7.png", std_array[7])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/std_8.png", std_array[8])
plt.imsave("/Users/Varun/Documents/Pattern_Recognition/Project1/Output/std_9.png", std_array[9])

covar_matrix = np.zeros((10, 784,784), dtype=np.float64)
for i in range(0,10):
    for key, value in enumerate(variance_arr[i]):
        covar_matrix[i][key][key] = value + 0.0001

y_pred = []
for i in X_test:
    final = []
    for j in range(10):
        inv = np.linalg.inv(covar_matrix[j])
        Wi = -0.5* inv
        Ni = np.matmul(mean_array[j].ravel(), inv)
        _, determinant = np.linalg.slogdet(covar_matrix[j])
        Bi = -0.5*np.matmul(np.matmul(mean_array[j].ravel().T, inv), mean_array[j].ravel()) + np.log(0.1)
        final.append(np.matmul(np.matmul(i.T, Wi), i) + np.matmul(Ni, i) + Bi)
    y_pred.append(np.argmax(np.array(final)))
        

loss = 0.0
for i, val in enumerate(y_pred):
    if val != y_test[i]:
        loss +=1
print("0-1 Loss is : {}".format((loss/len(y_pred))*100))
print("Accuracy is  {}".format(100 - (loss/len(y_pred))*100))
```
